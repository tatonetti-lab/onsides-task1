```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from glob import glob
from tqdm import tqdm
import re
#change to appropriate home folder
folder = '/content/drive/MyDrive/pop_pharmacogenomics/gpt/'
```

```python
#read in meddra vocabulary, and make dictionary
meddra_df = pd.read_csv(folder+'umls_meddra_en.csv')
meddra_df = meddra_df[['TTY', 'CODE', 'STR']].drop_duplicates()
meddra_df['STR'] = meddra_df.STR.apply(lambda x: x.lower())
meddra_dict = dict(zip(meddra_df.STR, meddra_df.CODE))
print(meddra_df.shape[0])
meddra_df.head(1)
```

```python
train_gpt35_df = pd.read_csv(folder+'train_drug_label_gpt35_output.csv')
mapped_ade_terms = []
for i, row in tqdm(train_gpt35_df.iterrows()):
  drug_name = row['drug_name']
  section_name = row['section_name']
  output = row['gpt35_output']
  candidate_terms = [i.lower() for i in re.split(r', |\n', output)]
  for term in candidate_terms:
      term = term.lower().strip()
      if term in meddra_dict.keys():
          mapped_ade_terms.append([drug_name, section_name, term, meddra_dict[term]])
      else:
          mapped_ade_terms.append([drug_name, section_name, term, None])
mapped_ade_terms = pd.DataFrame(mapped_ade_terms, columns=['drug_name', 'section_name', 'term', 'meddra_code'])
mapped_ade_terms.to_csv(folder+'train_drug_label_gpt35_ade_terms.csv', index=False)
mapped_ade_terms.head(1)
```
